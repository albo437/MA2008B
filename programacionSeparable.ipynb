# Tarea programación separable

```python
import pyomo.environ as pyo
from pyomo.opt import SolverFactory
```

# Problema 1

## Solución sobre problema original

```python
# Modelo de optimización no lineal
model = pyo.ConcreteModel()

# Variables de decisión 
model.X1 = pyo.Var(domain = pyo.NonNegativeReals)
model.X2 = pyo.Var(domain = pyo.NonNegativeReals)

# Función objetivo: 
model.obj = pyo.Objective(expr = pyo.exp(-model.X1) + model.X1 + (model.X2-1)**2, sense=pyo.maximize)

# Restricciones
model.con1 = pyo.Constraint(expr = model.X1**2 + model.X2 <= 3)

# Resolver el modelo
solver = SolverFactory('ipopt')
results = solver.solve(model)

print("x1 = " + str(model.X1.value))
print("x2 = " + str(model.X2.value))

print(f"z: {model.obj.expr()}")
```

## Solución sobre problema linealizado

```python
# Modelo de optimización no lineal
model = pyo.ConcreteModel()

# Variables de decisión 
model.X11 = pyo.Var(domain = pyo.NonNegativeReals)
model.X12 = pyo.Var(domain = pyo.NonNegativeReals)
model.X13 = pyo.Var(domain = pyo.NonNegativeReals)
model.X14 = pyo.Var(domain = pyo.NonNegativeReals)

model.X21 = pyo.Var(domain = pyo.NonNegativeReals)
model.X22 = pyo.Var(domain = pyo.NonNegativeReals)
model.X23 = pyo.Var(domain = pyo.NonNegativeReals)
model.X24 = pyo.Var(domain = pyo.NonNegativeReals)



# Función objetivo: 
model.obj = pyo.Objective(expr = model.X11 + 1.139*model.X12 + 1.469*model.X13 + 1.909*model.X14
                                + model.X21  + model.X23 + 4*model.X24, sense=pyo.maximize)

# Restricciones
model.con1 = pyo.Constraint(expr = 1/3*model.X12 + 4/3*model.X13 + 3*model.X14
                                + model.X22 + 2*model.X23 + 3*model.X24 <= 3)

model.con2 = pyo.Constraint(expr = model.X11 + model.X12 + model.X13 + model.X14 == 1)
model.con3 = pyo.Constraint(expr = model.X21 + model.X22 + model.X23 + model.X24 == 1)

# Resolver el modelo
solver = SolverFactory('glpk')
results = solver.solve(model)

print("x1 = " + str(0.577*model.X12.value + 1.1547*model.X13.value + 1.732*model.X14.value))
print("x2 = " + str(model.X22.value + 2*model.X23.value + 3*model.X24.value))

print(f"z: {model.obj.expr()}")
```

# Problema 2

## Solución problema original

```python
# Modelo de optimización no lineal
model = pyo.ConcreteModel()

# Variables de decisión 
model.X1 = pyo.Var(domain = pyo.NonNegativeReals)
model.X2 = pyo.Var(domain = pyo.NonNegativeReals)
model.X3 = pyo.Var(domain = pyo.NonNegativeReals)

# Función objetivo: 
model.obj = pyo.Objective(expr = model.X1**2 + pyo.exp(model.X2) + pyo.log(1 + model.X3), sense=pyo.minimize)

# Restricciones
model.con1 = pyo.Constraint(expr = model.X1 + model.X2 + model.X3 >= 3)
model.con2 = pyo.Constraint(expr = model.X1 <= 2)
model.con3 = pyo.Constraint(expr = model.X2 <= 1)
model.con4 = pyo.Constraint(expr = model.X3 <= 1)


# Resolver el modelo
solver = SolverFactory('ipopt')
results = solver.solve(model)

print("x1 = " + str(model.X1.value))
print("x2 = " + str(model.X2.value))
print("x3 = " + str(model.X3.value))

print(f"z: {model.obj.expr()}")
```
