```python
# Modelo de optimización no lineal
model = pyo.ConcreteModel()
# Variables de decisión
model.R = pyo.Var(domain=pyo.NonNegativeReals) # Libras de caucho
model.O = pyo.Var(domain=pyo.NonNegativeReals) # Libras de aceite
model.C = pyo.Var(domain=pyo.NonNegativeReals) # Libras de carbon

# Función objetivo: Minimizar costo
model.obj = pyo.Objective(expr=4*model.R + 1*model.O + 7*model.C, sense=pyo.minimize)

# Restricciones
model.con1 = pyo.Constraint(expr=model.R + model.O + model.C == 100) # Restricción de cantidad total
model.con2 = pyo.Constraint(expr=model.R >= 25) # Restricción de caucho
model.con3 = pyo.Constraint(expr=model.R <= 60) # Restricción de caucho
model.con4 = pyo.Constraint(expr=model.C >= 50) # Restricción de carbon
model.con5a = pyo.Constraint(expr=34 + 0.1 * model.R + 0.06 * model.O - 0.3 * model.C + 0.001 * model.R * model.O + 0.005 * model.O**2 + 0.001 * model.C**2 >=25)
model.con5b = pyo.Constraint(expr=34 + 0.1 * model.R + 0.06 * model.O - 0.3 * model.C + 0.001 * model.R * model.O + 0.005 * model.O**2 + 0.001 * model.C**2 <=35)
model.con6 = pyo.Constraint(expr=17 + 0.35 * model.R - 0.4 * model.O - 0.002 * model.R**2 >=16)
model.con7 = pyo.Constraint(expr=12.5 - 0.1 * model.O - 0.001 * model.O**2 >=12)

# Resolver el modelo
solver = SolverFactory('ipopt')
results = solver.solve(model)

# Mostrar resultados
print(f"Cantidad de caucho (R): {model.R.value} libras")
print(f"Cantidad de aceite (O): {model.O.value} libras")
print(f"Cantidad de carbon (C): {model.C.value} libras")
print(f"Costo total: {model.obj.expr()}")

# Verificar restricciones (valores de las restricciones)
print("\nVerificación de restricciones:")
print(f"R + O + C = {model.R.value + model.O.value + model.C.value}")
print(f"R >= 25: {model.R.value}")
print(f"R <= 60: {model.R.value}")
print(f"C >= 50: {model.C.value}")
print(f"Con5a: {34 + 0.1 * model.R.value + 0.06 * model.O.value - 0.3 * model.C.value + 0.001 * model.R.value * model.O.value + 0.005 * model.O.value**2 + 0.001 * model.C.value**2 }")
print(f"Con5b: {34 + 0.1 * model.R.value + 0.06 * model.O.value - 0.3 * model.C.value + 0.001 * model.R.value * model.O.value + 0.005 * model.O.value**2 + 0.001 * model.C.value**2}")
print(f"Con6: {17 + 0.35 * model.R.value - 0.4 * model.O.value - 0.002 * model.R.value**2}")
print(f"Con7: {12.5 - 0.1 * model.O.value - 0.001 * model.O.value**2}")
```

```python
import pyomo.environ as pyo
from pyomo.opt import SolverFactory

model = pyo.ConcreteModel()

model.X = pyo.Var(domain=pyo.NonNegativeReals)
model.Y = pyo.Var(domain=pyo.Reals)

model.obj = pyo.Objective(
    expr= model.X**2 + model.Y**2,
    sense=pyo.minimize
)

model.con1 = pyo.Constraint(expr=model.X + model.Y >=1)

solver = SolverFactory('lancelot')
results = solver.solve(model)

print(f"X = {pyo.value(model.X)}")
print(f"Y = {pyo.value(model.Y)}")
print(f"Costo total = {pyo.value(model.obj)}")
```

```python
import pyomo.environ as pyo
from pyomo.opt import SolverFactory

model = pyo.ConcreteModel()

model.X = pyo.Var(domain=pyo.NonNegativeReals)
model.Y = pyo.Var(domain=pyo.Reals)

model.obj = pyo.Objective(
    expr= pyo.exp(model.X) + (model.Y)**2,
    sense=pyo.minimize
)

model.con1 = pyo.Constraint(expr=model.X + model.Y <=4)
model.con2 = pyo.Constraint(expr=model.X**2 + model.Y >=3)

solver = SolverFactory('ipopt')
results = solver.solve(model)

print(f"X = {pyo.value(model.X)}")
print(f"Y = {pyo.value(model.Y)}")
print(f"Costo total = {pyo.value(model.obj)}")
```

```python
import numpy as np

best = None

for x in np.linspace(0, 4, 10000):
    ymin = 3 - x**2

    for y in np.linspace(ymin, 4 - x, 10000):
        obj = np.exp(x) + y**2

        if best is None or obj < best[0]:
            best = (obj, x, y)

print(best)
```
